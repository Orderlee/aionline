```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
import os

import imblearn
from imblearn.over_sampling import RandomOverSampler 
from sklearn.ensemble import GradientBoostingClassifier
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import KFold , cross_val_score
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV
import lightgbm as lgb
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/train.csv')#,index=False)
df
```

```python
# 옥외누수(out), 옥내누수(in), 정상(normal), 전기/기계음(noise), 환경음(other)

df['class']=0
df.loc[df['leaktype'] == 'out','class']=0
df.loc[df['leaktype'] == 'in','class']=1
df.loc[df['leaktype'] == 'normal','class']=2
df.loc[df['leaktype'] == 'noise','class']=3
df.loc[df['leaktype'] == 'other','class']=4
df
```

```python
X_smote = np.expand_dims(X_smote, axis=-1)
```

```python
skf = KFold(n_splits = 5, shuffle = True, random_state=100)
```

```python
for train , validation in skf.split(X_smote, y_smote):
  print('train valid rate:', len(train),len(validation))
```

```python
import keras
import tensorflow as tf
from keras.models import Sequential
from keras.layers import LSTM,Dense, Dropout, Activation
#from keras.optimizers import SGD
from keras.wrappers.scikit_learn import KerasClassifier
from keras.utils import np_utils
from sklearn.model_selection import cross_val_score
from keras.callbacks import ModelCheckpoint
from sklearn.model_selection import train_test_split
from keras.callbacks import EarlyStopping



# x = df[df.columns[1:].to_list()]
# y = df[df.columns[-1:].to_list()]

# train_test_split
#x_train, x_test, y_train, y_test = train_test_split(X_smote[train], y_smote[train] , test_size=0.2, shuffle=True, stratify=None, random_state=100)


#train, val = train_test_split(df, test_size=0.2, random_state=100)

#print(train.shape[0])

model = tf.keras.models.Sequential()

model.add(tf.keras.Input(shape=X_smote[train].shape[1:])) 
model.add(LSTM(units = 10, return_sequences = True))

model.add(LSTM(units = 10, return_sequences = True))

model.add(LSTM(units = 10, return_sequences = True))

model.add(LSTM(units = 10, return_sequences = True))

model.add(LSTM(units = 10))

model.add(Dense(units = 5,activation='softmax'))
model.summary()


#print(train)
#print(val)
```

```python
#filepath="/Users/woorethane/test/baseline/train/weights-improvement-{epoch:02d}.ckpt"#-{val_acc:.2f}
#filepath="/Users/woorethane/test/baseline/train/oversampling_kfold-{epoch:02d}.h5"
filepath="/content/drive/MyDrive/Colab Notebooks/oversampling_kfold-{epoch:02d}.h5"
checkpoint = ModelCheckpoint(filepath, monitor='val_loss', verbose=2, save_best_only=True, mode='min')
callbacks_list = [checkpoint]
```

```python
# x_train.shape, y_train.shape
X_smote[train].shape, y_smote[train].shape
```

```python
#model.load_weights('/content/drive/MyDrive/Colab Notebooks/baseline/LSTM4ray(uniy=50),dense=5_oversampling_.h5')
```

```python
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy',metrics = ['accuracy']) #'sparse_categorical_crossentropy' categorical_crossentropy
# early_stopping = EarlyStopping() # 조기종료 콜백함수 정의
history = model.fit(X_smote[train],
                    y_smote[train],
                    batch_size=16,
                    epochs=1,
                    validation_steps=5,
                    validation_data=(X_smote[validation], y_smote[validation]), #x_test, y_test
                    validation_batch_size=16,
                    verbose=2,
                    callbacks=callbacks_list) #x_train,y_train,

```
