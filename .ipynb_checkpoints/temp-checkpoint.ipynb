```python
import pandas as pd
from sklearn.model_selection import train_test_split
import csv
import numpy as np
from statistics import mean

# 주어진 train.csv 파일을 불러옵니다
df = pd.read_csv('train.csv')
# train dataset 과 validation dataset 으로 나눠줍니다. train data는 80%이고, validation data는 나머지 20% 입니다. 데이터 선택은 random으로 해줍니다.
train, val = train_test_split(df, test_size=0.2, random_state=100)

print(f"No. of training examples: {train.shape[0]}")

rows = train.values.tolist()
print(len(rows[0]))

in_data = []
out_data = []
normal_data = []
noise_data = []
other_data = []

out_total_list = []
in_total_list = []
normal_total_list = []
noise_total_list = []
other_total_list = []

for row in rows:
    if row[0] == 'out':
        out_data.append(row)
        out_total_list.extend(row[1:])
    elif row[0] == 'in':
        in_data.append(row)
        in_total_list.extend(row[1:])
    elif row[0] == 'normal':
        normal_data.append(row)
        normal_total_list.extend(row[1:])
    elif row[0] == 'noise':
        noise_data.append(row)
        noise_total_list.extend(row[1:])
    elif row[0] == 'other':
        other_data.append(row)
        other_total_list.extend(row[1:])

print("len(out_data)",len(out_data))
print(type(out_data[0][1]))
print("len(in_data)",len(in_data))
print("len(normal_data)",len(normal_data))
print("len(noise_data)",len(noise_data))
print("len(other_data)",len(other_data))
print("total length:",len(out_data) + len(in_data) + len(normal_data) + len(noise_data) + len(other_data))
```
